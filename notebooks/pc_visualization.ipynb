### Load data and visualize point cloud

```python
from src.data.data_module import load_dataset

raw_data = load_dataset()

cloud = raw_data[0].x
print(cloud)
```

```python
import plotly.graph_objects as go
import numpy as np

# Your [200,3] tensor
pc = cloud.cpu().numpy()  # Ensure numpy

# Colors & indices as arrays
colors = np.array(['blue'] * 100 + ['red'] * 100)
customdata = np.arange(200)

fig = go.Figure(data=[go.Scatter3d(
    x=pc[:,0], y=pc[:,1], z=pc[:,2],
    mode='markers',
    marker=dict(size=4, color=colors, line=dict(width=0.5, color='black'), opacity=0.8),
    name='Left (blue) / Right (red)',
    hovertemplate='<b>Point %{customdata}</b><br>X: %{x:.3f}<br>Y: %{y:.3f}<br>Z: %{z:.3f}<extra></extra>',
    customdata=customdata  # ✅ np.array
)])

fig.update_layout(
    title="Interactive Brain Point Cloud (Fixed)",
    scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z', aspectmode='cube')
)
fig.show()
```
